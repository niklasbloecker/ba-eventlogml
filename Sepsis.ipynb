```python
import pandas as pd
import matplotlib
```

```python
df = pd.read_csv('logs/sepsis.csv')
cases = df.case.unique()
```

```python
case_len = []
for i in cases:
    case_len.append([i,len(df[df.case == i])])

case_lendf = pd.DataFrame(case_len)
```

```python
case_lendf.plot() # Keine einheitliche LÃ¤nge der Traces
```

```python
print("Mean length: ", case_lendf[1].mean())
print("Max length: ", case_lendf[1].max())
print("Min length: ", case_lendf[1].min())
```
